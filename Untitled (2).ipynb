```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import yfinance as yf
```

```python
data = yf.download("AAPL", start = "2018-01-01")
data.head()
```

```python
mean = 0.001106
volatility = 0.019358
days = 252
simulations = 10000
starting_value = 50000

results = []

for i in range(simulations):
    random_returns = np.random.normal(mean, volatility, days)
    final_value = starting_value * np.prod(1 + random_returns)
    results.append(final_value)

var_95 = np.percentile(results , 5)

print("95% VaR:", starting_value - var_95)
```

```python
daily_returns = data["Close"].pct_change()
daily_returns = daily_returns.dropna() #first value has no previous day to compare to. So dropna to remove missing data
mean_returns = daily_returns.mean()
volatility = daily_returns.std()

print(mean_returns)
print(volatility)
```

```python
plt.hist(results)
plt.xlabel("Final Value")
plt.ylabel("Frequency")
plt.title("Monte Carlo Simulation -- APPL")
plt.show()
```

```python
#calculating win probability for APPL
count = 0
for final_value in results: 
    if final_value > starting_value:
        count +=1

win_probability = (count / simulations) * 100
print(f"Win Probability: {win_probability}%")
```

```python
#calculating CVaR
count_of_worst = 0
sum_of_worst_cases = 0

for final_value in results:
    if final_value <= var_95:
        sum_of_worst_cases += final_value
        count_of_worst += 1

average_bad_outcome = sum_of_worst_cases / count_of_worst



print(f"Average of worst 5% outcomes: {starting_value - average_bad_outcome}")


 

```

```python
#calculating average win
count_of_best = 0
sum_of_best_cases = 0 

for final_value in results:
    if final_value >= 50000:
        sum_of_best_cases += final_value
        count_of_best += 1

average_good_outcome = sum_of_best_cases / count_of_best


print(f"Average win: {average_good_outcome - starting_value}")
```

```python
spy_mean = 0.000534
spy_volatility = 0.012269
spy_days = 252
spy_simulations = 10000
spy_starting_value = 50000

spy_results = []

for i in range(spy_simulations):
    spy_random_returns = np.random.normal(spy_mean, spy_volatility, spy_days)
    spy_final_value = spy_starting_value * np.prod(1 + spy_random_returns)
    spy_results.append(spy_final_value)

spy_var_95 = np.percentile(spy_results , 5)

print("95% VaR:", spy_starting_value - spy_var_95)
```

```python
plt.hist(spy_results)
plt.xlabel("final value")
plt.ylabel("frequency")
plt.title("S&P Monte Carlo Simulation")
plt.show()
```

```python
spy_count = 0
for spy_final_value in spy_results:
    if spy_final_value > spy_starting_value:
        spy_count +=1

spy_win_probability = (spy_count / spy_simulations) * 100
print(f"S&P Win Probability: {spy_win_probability}%")
```

```python
spy_count_of_worst = 0
spy_sum_of_worst_cases = 0

for spy_final_value in spy_results:
    if spy_final_value <= spy_var_95:
        spy_sum_of_worst_cases += spy_final_value
        spy_count_of_worst += 1

spy_average_bad_outcome = spy_sum_of_worst_cases / spy_count_of_worst



print(f"S&P Average of worst 5% outcomes (95% CVAR): {spy_starting_value - spy_average_bad_outcome}")
```

```python
spy_count_of_best = 0
spy_sum_of_best_cases = 0 

for spy_final_value in spy_results:
    if spy_final_value >= 50000:
        spy_sum_of_best_cases += spy_final_value
        spy_count_of_best += 1

spy_average_good_outcome = spy_sum_of_best_cases / spy_count_of_best

print(f"S&P Win Probability: {spy_win_probability}%")
print(f"S&P Average win: {spy_average_good_outcome - spy_starting_value}")
print(f"S&P Average of worst 5% outcomes: {spy_starting_value - spy_average_bad_outcome}")
```
